```python
import tensorflow as tf 
import math

def conv(x_):
    conv1 = tf.layers.conv2d(
            inputs=x_,
            filters=32,  # number of filters
            kernel_size=[5, 5],
            padding="same",
            activation=tf.nn.relu)
    
    pool1 = tf.layers.max_pooling2d(inputs=conv1, 
                                    pool_size=[2, 2], 
                                    strides=2)  # convolution stride
    
    conv2 = tf.layers.conv2d(
            inputs=pool1,
            filters=64, # number of filters
            kernel_size=[5, 5],
            padding="same",
            activation=tf.nn.relu)
    
    pool2 = tf.layers.max_pooling2d(inputs=conv2, 
                                    pool_size=[2, 2], 
                                    strides=2)  # convolution stride
    conv3 = tf.layers.conv2d(
            inputs=pool2,
            filters=128, # number of filters
            kernel_size=[5, 5],
            padding="same",
            activation=tf.nn.relu)
    
    pool3 = tf.layers.max_pooling2d(inputs=conv3, 
                                    pool_size=[2, 2], 
                                    strides=2)  # convolution stride
    conv4 = tf.layers.conv2d(
            inputs=pool3,
            filters=256, # number of filters
            kernel_size=[5, 5],
            padding="same",
            activation=tf.nn.relu)
    
    pool4 = tf.layers.max_pooling2d(inputs=conv4, 
                                    pool_size=[2, 2], 
                                    strides=1)  # convolution stride
    conv5 = tf.layers.conv2d(
            inputs=pool4,
            filters=512, # number of filters
            kernel_size=[5, 5],
            padding="same",
            activation=tf.nn.relu)
    
    pool5 = tf.layers.max_pooling2d(inputs=conv5, 
                                    pool_size=[2, 2], 
                                    strides=1)  # convolution stride
    
    pool_flat = tf.contrib.layers.flatten(pool5, scope='pool2flat')
    dense = tf.layers.dense(inputs=pool_flat, units=5000, activation=tf.nn.relu)
    logits = tf.layers.dense(inputs=dense, units=3755)
    return logits


def classification(model_function):
    with tf.Graph().as_default() as g:
        with tf.device("/cpu:0"):  # use gpu:0 if on GPU
            x_ = tf.placeholder(tf.float32, [None, 58, 58, 3])
            drop_out = tf.placeholder(tf.float32)
            y_logits = model_function(x_)
            y_pred = tf.argmax(tf.nn.softmax(y_logits), axis=1)
    model_dict = {'graph': g, 'inputs': [x_], 'prediction': y_pred}
    
    return model_dict

def get_pred(xte):
    with model_dict['graph'].as_default(), tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        saver = tf.train.Saver()
        saver.restore(sess, 'checkpoints3755/checkpoint.ckpt')
        sen_s = []
        for i in range(len(xte)):
            sen_data = xte[i]
            test_feed_dict = dict(zip(model_dict['inputs'], [sen_data]))
            to_compute = [model_dict['prediction']]       
            result = sess.run(to_compute, test_feed_dict)
            sen_s.append(result)
        return sen_s
    
model_dict = classification(conv)
sen_s = get_pred(sen_data)
print('done')
```
